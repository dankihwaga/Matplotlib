```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line graph

```python
df = pd.read_csv('gas_prices.csv')
```

```python
plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 15})

plt.plot(df.Year, df.USA, 'b.-', label='USA')
plt.plot(df.Year, df.Canada, 'r.-', label='Canada')
plt.plot(df.Year, df['South Korea'], 'g.-', label='South Korea')
plt.plot(df.Year, df.Australia, 'y.-', label='Australia')

# Another Way to plot many values!
# countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(df.Year, df[country], marker='.')

plt.xticks(df.Year[::2])

plt.xlabel('Year', fontdict={'fontweight':'bold'})
plt.ylabel('US Dollars', fontdict={'fontweight':'bold'})

plt.legend()

#plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

### Histogram

```python
df2 = pd.read_csv('fifa_data.csv')
```

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(df2.Overall, bins=bins, color='#0b870f')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

#plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie chart

```python
left = df2.loc[df2['Preferred Foot'] == 'Left'].count()[0]
right = df2.loc[df2['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#cf4d34', '#12b0b5']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

df2.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in df2.Weight]

light = df2.loc[df2.Weight < 150].count()[0]
medium = df2[(df2.Weight >= 150) & (df2.Weight < 175)].count()[0]
medium_heavy = df2[(df2.Weight >= 175) & (df2.Weight < 200)].count()[0]
heavy = df2[df2.Weight >= 200].count()[0]

weights = [light, medium, medium_heavy, heavy]
label = ['under 150', '150-175', '175-200', 'over 200']

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = df2.loc[df2.Club == "FC Barcelona"]['Overall']
madrid = df2.loc[df2.Club == "Real Madrid"]['Overall']
manutd = df2.loc[df2.Club == "Manchester United"]['Overall']

#bp = plt.boxplot([barcelona, madrid, manutd], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, manutd], labels=['FC Barcelona','Real Madrid','Manchester United'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
